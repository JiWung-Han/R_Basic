### 1. Generate Random Number

```r
# set random seed
set.seed(77)

dataset = rt(200, df = 10)
```

```r
# 난수 생성 완료
dataset
```

### 2. Histogram


```r
# 히스토그램 생성

histo = hist(dataset, main="Histogram", col=0:9, labels=TRUE)
```

```r
histo
```

### 3. Box Plot

```r
# 상자그림 생성
boxp = boxplot(dataset, col=c("green"), main="Box plot")
```

```r
boxp
```

```r
median = 0.04588139

iqr = 0.76265025 -(-0.54964228)
conf_down = median - (1.58*iqr/sqrt(200))
conf_up = median + (1.58*iqr/sqrt(200))

sprintf("iqr : %f", iqr)
sprintf("conf_down : %f", conf_down)
sprintf("conf_up : %f", conf_up)
```

conf는 중앙값의 95% 신뢰구간을 나타낸다 <br>
계산법은 median + (1.58 * IQR / n^(1/2)) 과 median - (1.58 * IQR / n^(1/2)) 이다.
